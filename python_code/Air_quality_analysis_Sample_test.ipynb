```python
%matplotlib inline
import pandas as pd                #  data store and treat package pandas 
import matplotlib as mpl           # graph package
import matplotlib.pyplot as plt    # graph package
```

```python
Co_information = pd.read_csv('Indiana_co_data.csv',encoding='euc-kr')
Co_information
```

```python
Co_information['Date']
```

```python
Co_information['Daily Max 8-hour CO Concentration']
```

```python
# Date information 

Dates = []

# Co information

Cos = []

for x in range(len(Co_information['Date'])):
    Dates.append(Co_information['Date'][x])
print(Dates)
```

```python
for x in range(len(Co_information['Daily Max 8-hour CO Concentration'])):
    Cos.append(Co_information['Daily Max 8-hour CO Concentration'][x])
print(Cos)
```

```python
print(Cos[1037])
print(Co_information['Daily Max 8-hour CO Concentration'][1037])
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

plt.plot(Cos, color='green', marker='o', linestyle='solid')

plt.title("Indiana Outdoor Airquality CO")

plt.ylabel("ppm")
plt.show()
```
